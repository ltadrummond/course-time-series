```python
df = pd.read_csv('/home/leo/Projects_personal/course-time-series/data/retail_daily.csv')
df = df.copy()
print(df.info(), '\n\n')
df['date'] = pd.to_datetime(df['date'])
print(df.info())
```

```python
print("Duplicated status: ", df.duplicated('date').sum())
df.head()
```

```python
df = df.sort_values('date').set_index('date')
df
```

```python
print("timezone: ", df.index.tz)
```

```python
assert df.index.is_monotonic_increasing
print("Inferred frequency: ", pd.infer_freq(df.index))
```

```python
print("invalid negative: ", (df['price'] < 0).sum())
```

```python
df3 = df.rename(columns={'demand': 'y', 'date': 'timestamp'})
```

```python
full_df = df3['y'].asfreq('D')
int(full_df.isna().sum())
```
